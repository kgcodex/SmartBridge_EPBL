# Average Temperature Change Over Time

```python
# Synthetic Data: Simulate average yearly temperatures over 100 years
years = np.arange(1924, 2024) # 100 years from 1924 to 2023
base_temp = 15  # Base average temperature
temp_increase = np.linspace(0, 2, 100) # Gradual increase over 100 years
temp_fluctuations = np.random.normal(0, 0.5, 100) # Random yearly fluctuations
average_temps = base_temp + temp_increase + temp_fluctuations

# Plotting
plt.figure(figsize=(10, 6)) # Adjust figure size as needed
plt.plot(years, average_temps, label='Average Temperature (°C)')
plt.title('Average Temperature Change Over the Last Century in City X')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.grid(True) # Add grid for better readability
plt.legend()
plt.show()
```

# Population Change Over Time

# Correlation Matrix of House Features

```python

# Synthetic Data: House features and sale price
np.random.seed(42)
num_houses = 100
size = np.random.uniform(1000, 3000, num_houses) # Size in sq ft
bedrooms = np.random.randint(2, 6, num_houses)
age = np.random.randint(5, 50, num_houses)
location_factor = np.random.uniform(0.5, 1.5, num_houses) # Location influence on price
price = 100000 + 150 * size + 30000 * bedrooms - 1000 * age + 50000 * location_factor + np.random.normal(0, 50000, num_houses)
price = np.clip(price, 50000, 1000000) # Clip price to a reasonable range

df_houses = pd.DataFrame({'Size': size, 'Bedrooms': bedrooms, 'Age': age, 'Price': price})
correlation_matrix = df_houses.corr() # Calculate correlation matrix

# Plotting using Seaborn for better aesthetics
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5) # annot shows values, cmap color scheme
plt.title('Correlation Matrix of House Features and Sale Price')
plt.show()
```

# Website Visits Frequency

```python

# Synthetic Data: Website visits frequency by hour of day and day of week
days_of_week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
hours_of_day = range(24)
visits_frequency = np.zeros((len(days_of_week), len(hours_of_day)))

# Simulate higher traffic during weekdays and daytime hours
for day_index, day in enumerate(days_of_week):
    for hour_index in hours_of_day:
        if day_index < 5: # Weekdays (Mon-Fri)
            if 9 <= hour_index <= 17: # Daytime (9am-5pm)
                visits_frequency[day_index, hour_index] = np.random.randint(50, 150) # Higher traffic
            else:
                visits_frequency[day_index, hour_index] = np.random.randint(10, 50) # Lower traffic
        else: # Weekends (Sat-Sun) - generally lower
            visits_frequency[day_index, hour_index] = np.random.randint(5, 30) # Lower traffic on weekends

df_visits = pd.DataFrame(visits_frequency, index=days_of_week, columns=[f"{h:02d}:00" for h in hours_of_day]) # DataFrame for heatmap

# Plotting
plt.figure(figsize=(10, 7))
sns.heatmap(df_visits, cmap="YlGnBu", annot=False,
            cbar_kws={'label': 'Website Visits Frequency'}) # Pass cbar_label within cbar_kws
plt.title('Website Visits Frequency by Hour of Day and Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

# Flower Species Petal and Sepal Dimensions

# CO2 Emissions vs. Engine Size by Vehicle Class

```python

vehicle_classes = ['Compact', 'SUV', 'Sedan', 'Truck']
co2_emissions = {}
engine_size = {}
np.random.seed(42)
for v_class in vehicle_classes:
    engine_size[v_class] = np.random.uniform(1.0, 5.0, 50) # Engine sizes vary
    co2_emissions[v_class] = 150 + 50 * engine_size[v_class] + np.random.normal(0, 40, 50) # CO2 related to engine size

df_vehicles = pd.DataFrame({
    'Vehicle Class': np.repeat(vehicle_classes, 50),
    'Engine Size (L)': np.concatenate(list(engine_size.values())),
    'CO2 Emissions (g/km)': np.concatenate(list(co2_emissions.values()))
})

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size (L)', y='CO2 Emissions (g/km)', hue='Vehicle Class', data=df_vehicles, palette='Set1', alpha=0.7) # or sns.relplot(x='Engine Size (L)', y='CO2 Emissions (g/km)', hue='Vehicle Class', data=df_vehicles, kind='scatter')
plt.title('CO2 Emissions vs. Engine Size by Vehicle Class (Seaborn Scatterplot)')
plt.show()
```
